```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 페이지를 구해주는 함수.
def findpagedata(url):
    from selenium import webdriver

    driver_path = 'chromedriver.exe'
    options = webdriver.ChromeOptions()
    options.add_argument("headless")
    
    driver = webdriver.Chrome(driver_path,options=options)
    
    driver.get(url)
    

    result = int(driver.find_element(By.XPATH,r'//*[@id="prdNoticeTable"]/tbody/tr[2]/td[2]/p').text.strip()[:-1])
    driver.close()
    return result
```

```python
import requests

url = 'http://book.interpark.com/api/search.api'
params ={'key' : 'D800D854380C5E24CE00448963946732F9F363BA05BBBE07C8895E035504DBC6', 'query':input("시 분야 검색"),'output':'json','maxResults':'100','categoryId':'102'}

response = requests.get(url, params=params)

result = str(response.content,'utf-8')
```

```python
import json

js = json.loads(result)
```

```python
booklist = js['item']
```

```python
booklist
```

```python
poempage = []
poemprice = []
```

```python
computerpage = []
computerprice = []
```

```python
for i in range(len(booklist)):
    try:
        data1 = findpagedata(booklist[i]['link'])
        data2 = booklist[i]['priceStandard']

        computerpage.append(data1)
        computerprice.append(data2)
    except:
        continue
```
